```python
# SVM on an Artificially Generated Dataset
# Paste this into Google Colab

# 1. Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# 2. Generate synthetic data
def generate_dataset(n_per_class=100, noise=0.5, random_state=42):
    """
    Generates a 2-D synthetic dataset with two classes, partially overlapping.
    """
    np.random.seed(random_state)
    # Class 0: centered at (2,2)
    X0 = np.random.randn(n_per_class, 2) * noise + np.array([2, 2])
    y0 = np.zeros(n_per_class, dtype=int)
    # Class 1: centered at (7,7)
    X1 = np.random.randn(n_per_class, 2) * noise + np.array([7, 7])
    y1 = np.ones(n_per_class, dtype=int)
    # Combine
    X = np.vstack((X0, X1))
    y = np.concatenate((y0, y1))
    return X, y

# 3. Prepare data
X, y = generate_dataset(n_per_class=200, noise=1.0, random_state=0)
# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)

# 4. Train SVM classifier
model = svm.SVC(kernel='rbf', C=1.0, gamma='scale')  # you can change kernel to 'linear', 'poly', etc.
model.fit(X_train, y_train)

# 5. Predict & evaluate
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification report:")
print(classification_report(y_test, y_pred))

# 6. Plot decision boundary
def plot_decision_boundary(model, X, y, ax=None):
    if ax is None:
        fig, ax = plt.subplots(figsize=(8,6))
    # define grid
    x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
    y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 200),
        np.linspace(y_min, y_max, 200)
    )
    # predict on grid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # plot contour + classes
    ax.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    ax.scatter(X[:,0], X[:,1], c=y, s=40, edgecolors='k', cmap=plt.cm.coolwarm)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_title(f"SVM Decision Boundary (kernel={model.kernel})")
    return ax

plot_decision_boundary(model, X_train, y_train)
plt.show()
```
